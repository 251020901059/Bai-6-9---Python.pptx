```python
n=int(input("Enter a positive integer:"))
print(n)

s=0 # bien so dung de luu tong
# nhap code cua ban vao day

for i in range(n + 1):   # duyệt từ 0 đến n
    s += i               # cộng dồn vào biến

print(f"The sum of 1 to {n} is {s}")
```
